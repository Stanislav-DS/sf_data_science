# ЗАДАНИЕ 5.7 (ВЫПОЛНЯЕТСЯ В CODEBOARD НИЖЕ)

Напишите бесконечный итератор по списку.

Для этого создайте генератор с названием `inf_iter`, который принимает на вход список и возвращает элементы из него через `yield`.

Когда элементы в списке заканчиваются, генератор снова возвращает элементы из списка, начиная с нулевого.

Пример работы итератора:

    l = [101, 102, 103]
    gen = inf_iter(l)
    for _ in range(10):
        print(next(gen))

```python
def inf_iter(lst):
    while True:
        for elem in lst:
            yield elem

lst = [101, 102, 103]
gen = inf_iter(lst)
for _ in range(10):
    print(next(gen))
```

# ЗАДАНИЕ 5.8 (ВЫПОЛНЯЕТСЯ В CODEBOARD НИЖЕ)

Напишите генератор `group_gen(n)`. Он должен при каждом вызове выдавать порядковый номер от `1` до `n` (включая `n`). После достижения `n` генератор должен снова возвращать номера, начиная с `1`.

Пример работы функции:

    groups = group_gen(3)
    for _ in range(10):
        print(next(groups))
    # 1
    # 2
    # 3
    # 1
    # 2
    # 3
    # 1
    # 2
    # 3
    # 1

```python
def group_gen(n):
    while True:
        for i in range(1, n + 1):
             yield i


groups = group_gen(3)
for _ in range(10):
    print(next(groups))
# 1
# 2
# 3
# 1
# 2
# 3
# 1
# 2
# 3
# 1
```
