## ЗАДАНИЕ 3.4 (ВЫПОЛНЯЕТСЯ В CODEBOARD НИЖЕ)

Напишите рекурсивную функцию `multiply_lst()`, которая перемножает элементы заданного списка между собой. 

Переведите алгоритм создания рекурсивной функции для расчёта суммы в расчёт произведения. Условие остановки — `if len(lst) == 0: return 1`.

Мой код:

```python
def multiply_lst(lst):
    if not lst:
        return 1
    return lst[0] * multiply_lst(lst[1:])


my_lst = [10, 21, 24, 12]
print(multiply_lst(my_lst))
```

## ЗАДАНИЕ 3.7 (ВЫПОЛНЯЕТСЯ В CODEBOARD НИЖЕ)

Числа Фибоначчи — пример последовательности, которую можно получить рекурсивно. Вот первые числа этой последовательности: `0, 1, 1, 2, 3, 5, 8, 13 …`

Как уже отмечалось, последовательность Фибоначчи имеет важное значение в комбинаторике и аналитике.

Как вы могли заметить, каждый `n`-ый элемент — это сумма `n - 1` и `n - 2` элементов.

Можно записать рекурсивное выражение для расчёта `n`-ого числа из последовательности Фибоначчи:

2022-07-12_22-03-34.png

Напишите рекурсивную функцию `fib(n)`, которая считает `n`-ое число Фибоначчи. Будем считать, что `fib(0) = 0` и `fib(1) = 1`.

Пример работы функции:

    print(fib(0))
    print(fib(2))
    print(fib(6))
    # 0
    # 1
    # 8

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)


print(fib(0))
print(fib(2))
print(fib(6))
```
